```python
x = df.drop(['CLIENTNUM', 'Attrition_Flag'], axis=1)
y = df['Attrition_Flag'].values
```

```python
# scaler = StandardScaler().fit(x)
# x = scaler.transform(x)
```

```python
# import packages
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
pca = PCA(5)
pca.fit(x)
x =  pca.transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=111)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
r_model = RandomForestClassifier(n_estimators=500, n_jobs=-1)
```

```python
r_model.fit(x_train, y_train)
```

```python
print(r_model.score(x_train, y_train))
print(r_model.score(x_test, y_test))
```

```python
# logistic model
l_model = LogisticRegression(max_iter=500)
```

```python
l_model.fit(x_train, y_train)

print(l_model.score(x_train, y_train))
```

```python
l_model.score(x_test, y_test)
```
