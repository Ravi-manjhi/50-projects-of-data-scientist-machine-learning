```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import metrics
from sklearn.preprocessing import LabelEncoder
from dateutil import parser
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/content/drive/MyDrive/Data Science/Project-34 Google Play App Rating prediction/googleplaystore.csv')
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
# time data processing
 
s = "March 25, 2005"
p = parser.parse(s)
p.strftime("%D")
```

```python
def mapper(x):
  try:
    return pd.to_datetime(x)
  except:
    return np.nan


df['Last Updated'] = df['Last Updated'].apply(mapper)
```

```python
# size of apps

df['Size'] = df['Size'].replace('Varies with device', '15')
df['Size'] = df['Size'].apply(lambda x: x.replace('M', '') if 'M' in x else x.replace('k', ''))
df['Size'] = df['Size'].apply(lambda x: pd.to_numeric(x))
```

```python
def clean_install(x):
  num = [i for i in x if i.isnumeric()]
  return int(''.join(num))
```

```python
df['Installs'] = df['Installs'].apply(clean_install)
```

```python
# reviews

df['Reviews'] = pd.to_numeric(df['Reviews'])
```

```python
# type 

df['Type'].value_counts()
```

```python
df['Type'] = df['Type'].map({'Free': 0, 'Paid': 1})
```

```python
# Android Ver

df['Android Ver'] = df['Android Ver'].apply(lambda x: int(x.split('.')[0]) if '.' in x else 4)
```

```python
# Category
cat_encoder = LabelEncoder().fit(df['Category'])
df['Category'] = cat_encoder.transform(df['Category'])
```

```python
dict(zip(cat_encoder.classes_, cat_encoder.transform(cat_encoder.classes_)))
```

```python
# Content Rating

con_encoder = LabelEncoder().fit(df['Content Rating'])
df['Content Rating'] = con_encoder.transform(df['Content Rating'])
```

```python
dict(zip(con_encoder.classes_, con_encoder.transform(con_encoder.classes_)))
```

```python
df['Price'] = df['Price'].apply(lambda x: np.float64(x.replace('$', '')))
```

```python
df.head()
```

```python
df['Day'] = df['Last Updated'].apply(lambda x: x.day)
df['Month'] = df['Last Updated'].apply(lambda x: x.month)
df['Year'] = df['Last Updated'].apply(lambda x: x.year)
```

```python
app = df.drop(['App', 'Genres', 'Current Ver', 'Last Updated'], axis=1).copy()
```

```python
app.head()
```

```python
app = pd.get_dummies(data=app, columns=['Content Rating'], drop_first=True)
```

```python
plt.figure(figsize=(15, 10))
sns.heatmap(app.corr(), annot=True)
plt.show()
```

```python
x = app.drop(['Rating'], axis=1)
y = app['Rating'].values
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=5).fit(x)
x = pca.transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=111)
```

```python
x_train.shape, y_train.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rg = RandomForestRegressor().fit(x_train, y_train)
```

```python
rg.score(x_train, y_train)
```

```python
rg.score(x_test, y_test)
```
