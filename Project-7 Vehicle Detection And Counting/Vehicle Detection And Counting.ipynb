# vehical Detect-Counts

```python
import numpy as np
import pandas as pd
import requests
import cv2 as cv
from PIL import Image
```

```python
img = Image.open("./cars.png")
img = img.resize((450, 250))
img
```

```python
img_array = np.array(img)
```

```python
gray = cv.cvtColor(img_array, cv.COLOR_RGB2GRAY)
Image.fromarray(gray)
```

```python
blur = cv.GaussianBlur(gray, (5, 5), 0)
Image.fromarray(blur)
```

```python
# dilate the image
kernel = np.ones((3, 3))

dilated = cv.dilate(blur, kernel)
Image.fromarray(dilated)
```

```python
# Morphology tranformation

kernel = cv.getStructuringElement(cv.MORPH_ELLIPSE, (2, 2))
closing = cv.morphologyEx(dilated, cv.MORPH_CLOSE, kernel)

Image.fromarray(closing)
```

```python
car_cascade = cv.CascadeClassifier('./cars.xml')
cars = car_cascade.detectMultiScale(closing, 1.1, 1)
```

```python
# drawing rectangle

cnt = 0
for (x, y, w, h) in cars:
  cv.rectangle(closing, (x, y), (x+w, y+h), (255, 0, 0), 2)
  cnt += 1

print(cnt, 'cars founds')
Image.fromarray(img_array)
```

## Bus detecting

```python
def processing_img(path):
  img = Image.open(path)
  img_array = np.array(img)
  gray = cv.cvtColor(img_array, cv.COLOR_RGB2GRAY)
  blur = cv.GaussianBlur(gray, (5, 5), 0)
  dilated = cv.dilate(blur, np.ones((3, 3)))
  final_img = cv.morphologyEx(dilated, cv.MORPH_CLOSE, cv.getStructuringElement(cv.MORPH_ELLIPSE, (2, 2)))
  return img_array, final_img
```

```python
bus_array, bus_final = processing_img('./bus.jpg')
```

```python
bus_cascade = cv.CascadeClassifier('./bus.xml')
```

```python
bus = bus_cascade.detectMultiScale(bus_final, 1.1, 4)
```

```python
cnt = 0
for (x, y, w, h) in bus:
  cv.rectangle(bus_array, (x, y), (x+w, y+h), (0, 255, 0), 2)
  cnt += 1

print(cnt, 'cars founds')
Image.fromarray(bus_array)
```
