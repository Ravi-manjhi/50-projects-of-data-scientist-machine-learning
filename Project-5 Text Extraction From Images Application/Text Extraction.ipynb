```python
image = cv.imread('./01.png')
```

```python
def gray(img):
  return cv.cvtColor(img, cv.COLOR_RGB2GRAY)
```

```python
gray = gray(image)
```

```python
img = Image.fromarray(gray)
```

```python
# remove noice
def remove_noise(img):
  return cv.medianBlur(img, 5)

noise = remove_noise(image)
```

```python
# Image.fromarray(noise)
```

```python
# thresholding

def threshold(img):
  return cv.threshold(image, 0, 255, cv.THRESH_BINARY)[1]
```

```python
thresh = threshold(gray)
```

```python
Image.fromarray(thresh)
```

```python
# erosion
def erode(img):
  kernel = np.ones((5, 5), np.uint8)
  return cv.erode(img, kernel,1)
```

```python
img = erode(image)
```

```python
Image.fromarray(img)
```
