# Import Packages

```python
import pandas as pd
import numpy as np
import difflib
import neattext.functions as nfx
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity 
```

# Data preprocessing

```python
df = pd.read_csv('./udemy_tech.csv')
```

```python
df.head()
```

```python
df['Title'] = df['Title'].apply(nfx.remove_shortwords).apply(nfx.remove_special_characters)
```

```python
df['Title'][0]
```

```python
countvect = CountVectorizer()

cmat = countvect.fit_transform(df['Title'])
cmat
```

```python
cos_sim = cosine_similarity(cmat)
```

```python
cos_sim.shape
```

## Recommend Course

```python
title = '100 Days of Python'
```

```python
course_index = pd.Series(
    df.index, index=df['Title']
).drop_duplicates()


score = list(enumerate(cos_sim[2]))

sorted_score = sorted(score, key=lambda x: x[1], reverse=True)

selected_course_index = [index[0] for index in sorted_score[1: ]]

selected_course_score = [score[1] for score in sorted_score[1: ]]

rec_df = df.iloc[selected_course_index]

rec_df['Similarity_score'] = selected_course_score

final_recommendation = rec_df[['Title', 'Link']]
```
