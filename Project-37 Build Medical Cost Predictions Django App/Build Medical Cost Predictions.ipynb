```python
x = df.drop('charges', axis=1)
y = df['charges'].values
```

```python
log_y = np.log(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=55)
```

```python
all_models = [LinearRegression, Lasso, DecisionTreeRegressor, SVR, RandomForestRegressor, XGBRegressor, XGBRFRegressor]
```

```python
def getModelScore(model):
  model = model().fit(x_train, y_train)

  print(type(model).__name__)
  print("Training score",model.score(x_train, y_train))

  y_pred = model.predict(x_test)
  print("Testing score", model.score(x_test, y_test))
  print("Testing Score R2 Score", metrics.r2_score(y_test, y_pred))
  print("Testing Score MSE Score", metrics.mean_squared_error(y_test, y_pred))
  print("Testing Score MSE Score", metrics.mean_squared_error(y_test, y_pred))
  return model
```

```python
model_dict= {}
for i in all_models:
  model = getModelScore(i)
  model_dict[type(model).__name__] = model
  
  print('---'*20)
```
