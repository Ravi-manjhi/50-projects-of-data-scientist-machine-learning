## Splitting data

```python
# dependent and independent variable
x = df.drop(['Class'], axis=True)
y = df['Class'].values
```

```python
# splitting into training and testing dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=44)
```

## Isolation Forest

```python
iso = IsolationForest(n_estimators=200, max_samples=len(x_train), random_state=0, verbose=0)
```

```python
iso.fit(x_train, y_train)
```

```python
y_pred = iso.predict(x_test)
```

```python
np.unique(y_pred)
```

```python
# mapping the values
y_pred[y_pred == 1] = 0
y_pred[y_pred == -1] = 1
```

```python
# accuracy metrics

metrics.accuracy_score(y_test, y_pred)
```

```python
print(metrics.confusion_matrix(y_test, y_pred))
```
