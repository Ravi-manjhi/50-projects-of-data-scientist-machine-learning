```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler

# import models
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
df = pd.read_csv('/content/drive/MyDrive/Data Science/ Project-36 Build Artist Sculpture Cost Prediction /Artist Sculpture Cost.csv')
```

```python
# checking null values
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df = df[df['Cost'] > 0]
```

```python
df.describe()
```

```python
df.head()
```

```python
df['Delivery Date'] = pd.to_datetime(df['Delivery Date'])
df['Scheduled Date'] = pd.to_datetime(df['Scheduled Date'])
```

```python
object_col =  ['Material', 'International',
       'Express Shipment', 'Installation Included', 'Transport', 'Fragile',
       'Customer Information', 'Remote Location']
```

```python
for i in object_col:
  print(i)
  print(df[i].value_counts())
  print("--"*20)
```

```python
for i in object_col:
  encoder = LabelEncoder().fit(df[i])
  df[i] = encoder.transform(df[i])
  print(dict(zip(encoder.classes_, encoder.transform(encoder.classes_))))
```

```python
df['Waiting'] = (df['Scheduled Date'] - df['Delivery Date']).apply(lambda x: int(str(x).split(' ')[0]))
```

```python
df.head()
```

```python
x = df.drop(['Customer Id', 'Artist Name', 'Scheduled Date', 'Delivery Date',
            'Customer Location', 'Cost'], axis=1).values
y = df['Cost'].values
```

```python
# scaler = StandardScaler().fit(x)
# x = scaler.transform(x)
```

```python
y = np.log(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=111)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
lr_model = LinearRegression().fit(x_train, y_train)
```

```python
lr_model.score(x_train, y_train)
```

```python
lr_model.score(x_test, y_test)
```

```python
# Random forest regression

rf_model = RandomForestRegressor(n_estimators=300, n_jobs=-1).fit(x_train, y_train)
```

```python
rf_model.score(x_train, y_train)
```

```python
rf_model.score(x_test, y_test)
```
