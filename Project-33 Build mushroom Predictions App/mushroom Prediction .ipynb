```python
def label_encoder(feat):
  encoder = LabelEncoder()
  encoder.fit(feat)
  x = encoder.transform(feat)
  return x, encoder
```

```python
encoder_list = {}
for i in object_name:
  feat, encoder = label_encoder(df[i])
  df[i] = feat
  encoder_list[i] = dict(zip(encoder.classes_, encoder.transform(encoder.classes_)))
```

# Buidling the model

```python
# splitting data into dependent and independent

x = df.drop('class', axis=1)
y = df['class'].values
```

```python
pca = PCA(n_components=7).fit(x)
pca2 = PCA(n_components=15).fit(x)

x_1 = pca.transform(x)
x_2 = pca.transform(x)
```

```python
# splitting data 

x_train_1, x_test_1, y_train_1, y_test_1 = train_test_split(x_1, y, test_size=0.2, random_state=101)
x_train_2, x_test_2, y_train_2, y_test_2 = train_test_split(x_2, y, test_size=0.2, random_state=101)
```

```python
all_models = [LogisticRegression, DecisionTreeClassifier, RandomForestClassifier, SVC, XGBClassifier]
```

```python
def get_model_score(model, x, y):
  x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=101)

  model = model()
  model.fit(x_train, y_train)
  
  print(type(model).__name__)
  print('The Training data Score', model.score(x_train, y_train))
  print('The Testing data Score', model.score(x_test, y_test))

  return model
```

```python
model_dict = {}
for i in all_models:
  get_model_score(i, x_1, y)
  print('--'*20)
```
